```python
# Initiate a spark session - the way for Spark 2.0 instead of sparkcontext
# spark.sql.warehouse.dir is only required if using windows
spark = SparkSession.builder.config("spark.sql.warehouse.dir", "file:///C:/temp").appName("SparkSQL").getOrCreate()
```

```python
#Writing a mapper function to delimit the data
def mapper(line):
    fields = line.split(',')
    return Row(ID=int(fields[0]), name=str(fields[1].encode("utf-8")), age=int(fields[2]), numFriends=int(fields[3]))

lines = spark.sparkContext.textFile("file:///C:/SparkCourse/fakefriends.csv")
people = lines.map(mapper)
```
